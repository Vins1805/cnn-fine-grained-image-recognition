```python
import os
import cv2
import torch
import torch.optim
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
```

```python
test_size = 106/606
batch_size = 10
random_state = 0
shuffle = True

filepath = "datasets/CUB_200_2011/images/"
folders = os.listdir(filepath)[1:]
```

```python
class MacnnDataset(Dataset):
    def __init__(self, x, y):
        super(MacnnDataset, self).__init__()
        self.x = x
        self.y = y

    def __getitem__(self, index):
        img = torch.tensor(cv2.resize(self.x[index], (448, 448))).float().unsqueeze(0)
        label = self.y[index]
        return img, label

    def __len__(self):
        return len(self.x)
```

```python
x, y = [], []

for index, folder in enumerate(folders):
    for file in os.listdir(filepath + folder):
        img = cv2.imread(filepath + folders[0] + "/" + os.listdir(filepath + folders[0])[0])
        x.append(img)
        y.append(index)
```

```python
x, tx, y, ty = train_test_split(x, y, test_size=test_size, random_state=random_state)
```

```python
train_set = MacnnDataset(x, y)
train_loader = DataLoader(dataset=train_set, batch_size=batch_size, shuffle=shuffle)
test_set = MacnnDataset(tx, ty)
test_loader = DataLoader(dataset=test_set, batch_size=batch_size, shuffle=shuffle)
```
